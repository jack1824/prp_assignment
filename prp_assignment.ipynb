```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

STEP1: Data Preparation (Pandas) - Load CSV dataset with columns like: Order_ID,
Customer_ID, Product, Category, Quantity, Price, Order_Date. - Explore dataset using:
.head(), .tail(), .info(), .describe(), .value_counts(), .unique(), .nunique(). - Handle missing
values and duplicates: .fillna(), .dropna(), .drop_duplicates().

```python
df = pd.read_csv('dataset.csv')
print("Dataset loaded successfully!")
```

```python
print("Unique values in the 'Category' column:")
display(df['Category'].unique())

print("\nNumber of unique values in the 'Customer_ID' column:")
display(df['Customer_ID'].nunique())

print("\nNumber of unique values in each column of the DataFrame:")
display(df.nunique())
```

```python

print("Missing values per column:")
display(df.isnull().sum())

df_filled = df.copy() # Create a copy to avoid modifying the original DataFrame
df_filled['Quantity'] = df_filled['Quantity'].fillna(df_filled['Quantity'].mean())
print("\nDataFrame after filling missing 'Quantity' with the mean:")
display(df_filled.head())

# can lead to significant data losss
df_dropped_rows = df.copy()
df_dropped_rows.dropna(inplace=True)
print("\nDataFrame after dropping rows with any missing values:")
display(df_dropped_rows.head())

# dropping duplicate rows
df_no_duplicates = df.copy()
df_no_duplicates.drop_duplicates(inplace=True)
print("\nDataFrame after dropping duplicate rows:")
display(df_no_duplicates.head())
```

STEP2:Data Analysis (NumPy) - Convert relevant columns to NumPy arrays (1-D for
individual columns, 2-D for numeric features). - Calculate: - Total sales per product and
per category. - Average order value per customer. - Monthly sales totals. - Percentage
contribution of each category to total revenue.

```python
# Convert a single column to a 1-D NumPy array (e.g., 'Price')
price_array = df['Price'].values
print("1-D NumPy array for 'Price' column:")
display(price_array[:10]) # Displaying first 10 elements

# Convert multiple numeric columns to a 2-D NumPy array
# Select only the numeric columns
numeric_cols = df.select_dtypes(include=np.number)
numeric_array = numeric_cols.values

# Set NumPy print options to suppress scientific notation
np.set_printoptions(suppress=True)

print("\n2-D NumPy array for numeric columns:")
display(numeric_array[:5]) # Displaying first 5 rows

# Reset NumPy print options to default (optional, but good practice)
np.set_printoptions(suppress=False)
```

```python
# Ensure 'Order_Date' is in datetime format for monthly calculations
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

# Calculate Total Sales per Product
# We need to handle potential missing values in 'Quantity' and 'Price' for this calculation
# Let's fill missing 'Quantity' with 1 and missing 'Price' with 0 for sales calculation purposes, or drop rows with missing sales data
# Dropping rows with missing Quantity or Price for accurate sales calculation
df_sales = df.dropna(subset=['Quantity', 'Price']).copy()
df_sales['Total_Sales'] = df_sales['Quantity'] * df_sales['Price']
total_sales_per_product = df_sales.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
print("Total Sales per Product:")
display(total_sales_per_product)

# Calculate Total Sales per Category
total_sales_per_category = df_sales.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales per Category:")
display(total_sales_per_category)

# Calculate Average Order Value per Customer
# We need to group by Customer_ID and calculate the mean of Total_Sales for each customer
# First, ensure we are using the dataframe with calculated Total_Sales and valid Customer_ID
df_customer_sales = df_sales.dropna(subset=['Customer_ID']).copy()
average_order_value_per_customer = df_customer_sales.groupby('Customer_ID')['Total_Sales'].mean().sort_values(ascending=False)
print("\nAverage Order Value per Customer:")
display(average_order_value_per_customer.head()) # Displaying head as there can be many customers

# Calculate Monthly Sales Totals
# We need to extract the month from the 'Order_Date'
df_monthly_sales = df_sales.dropna(subset=['Order_Date']).copy()
df_monthly_sales['Order_Month'] = df_monthly_sales['Order_Date'].dt.to_period('M')
monthly_sales_totals = df_monthly_sales.groupby('Order_Month')['Total_Sales'].sum().sort_index()
print("\nMonthly Sales Totals:")
display(monthly_sales_totals)

# Calculate Percentage Contribution of each Category to Total Revenue
total_revenue = total_sales_per_category.sum()
percentage_contribution_per_category = (total_sales_per_category / total_revenue) * 100
print("\nPercentage Contribution of each Category to Total Revenue:")
display(percentage_contribution_per_category)
```

STEP3:Data Visualization - Create meaningful visualizations: 1. Line Graph: Monthly sales
trend over the year. 2. Bar Graph: Total sales per product category. 3. Pie Chart:
Contribution of top 5 products to total sales. 4. Histogram: Distribution of order values. 5.
Scatter Plot: Quantity vs Revenue per order.

```python
# 1. Line Graph: Monthly sales trend over the year
plt.figure(figsize=(12, 6))
monthly_sales_totals.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# 2. Bar Graph: Total sales per product category
plt.figure(figsize=(10, 6))
total_sales_per_category.plot(kind='bar')
plt.title('Total Sales per Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Pie Chart: Contribution of top 5 products to total sales
# top 5 products by total sales
top_5_products = total_sales_per_product.head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_5_products, labels=top_5_products.index, autopct='%1.1f%%', startangle=140)
plt.title('Contribution of Top 5 Products to Total Sales')
plt.axis('equal')
plt.show()
```

```python
# 4. Histogram: Distribution of order values
#  using 'Total_Sales' from df_sales
plt.figure(figsize=(10, 6))
sns.histplot(df_sales['Total_Sales'], bins=20, kde=True)
plt.title('Distribution of Order Values (Total Sales)')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Scatter Plot: Quantity vs Revenue per order
#  'Quantity' and 'Total_Sales'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_sales['Quantity'], y=df_sales['Total_Sales'])
plt.title('Quantity vs Total Sales per Order')
plt.xlabel('Quantity')
plt.ylabel('Total Sales')
plt.show()
```

### Peak Sales Months

*   The line graph of monthly sales trend indicates that **July 2025** had the highest sales volume during the observed period. There also appear to be peaks in **December 2024** and **June 2025**.

### Customer Buying Patterns

*   The average order value per customer varies, with some customers having significantly higher average order values, suggesting potential high-value customers.
*   The scatter plot of Quantity vs Total Sales per Order shows a general positive correlation, as expected, but there are variations indicating that price and product type significantly influence total sales value beyond just quantity.
